```
import torch
```

```
#Uninitialized matrix
x = torch.empty(2,4)
print(x)
```

```
# Randomly initialized matrix
x = torch. rand(2,2)
print(x)
```

```
# Zero Matrix
z = torch.zeros(5,2,dtype=torch.long)
print(z)
```

```
# Create tensor directly
data = torch.tensor([5.5,3])
print(data)
```

```
# Modify tensor to new tensor
z = z.new_ones(5,3,dtype = torch.long)
print(z)

w = torch.randn_like(z,dtype=torch.float)
print(w)
```

```
# Get size
print(w.size())
```

```
# Addition syntaxes
y = torch.rand(5,3)
print(z+y)

print(torch.add(y,z))

result = torch.empty(5,3)
torch.add(y,z,out=result)
print(result)
```

```
# Inplace addition
a = torch.zeros(5,3)
b = torch.ones(5,3)
b.add_(a)
print(b)
```

```
# Resizing with torch.view
a = torch.rand(4,4)
b = a.view(16)
c = a.view(-1,8)
print("{} {} {}".format(a.size(),b.size(),c.size()))
```

```
# tensor to numpy bridge
a = torch.ones(5)
b = a.numpy()
print("{} {}".format(a,b))

# Pass by reference caveat
a.add_(1)
print("{} {}".format(a,b))
```

```
# Numpy to tensor
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a)
print(a,b)

# Pass by reference caveat
np.add(a, 1, out=a)
print(a)
print(b)
```

```
if torch.cuda.is_available():
  device = torch.device()
  b = torch.ones_like(a)
  a = a.to(device)
  ans = a + b
  print(ans)
  print(ans.to("cpu",torch.double))
```
