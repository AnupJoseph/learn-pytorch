```
import torch
```

```
# Create a tensor with tracked gradient computation
x = torch.ones(3,3,requires_grad = True)
print(x)
```

```
y = x + 2
print(y)
print(y.grad_fn)
```

```
z = y * y * 3
out = z.mean()

print(z, out)
```

```
# Use requires_grad_ function attribute over a Tensor to compute gradients from a certain point
a = torch.randn(2, 2)
a = ((a*3)/(a-1))
print(a.requires_grad)

# set requires grad to true
a.requires_grad_(True)
print(a.requires_grad)

# Create a new variable using a .This variable will have a as its grad_fn
b = (a*a).sum()
print(b)
print(b.grad_fn)
```

```
out.backward()
print(x.grad)
```
