```
# Alternative way
# Kinda useful if you want to compute without gradients first and then add gradients later
# I don't know why you would do that

b = torch.ones((2,2))
print(b)
b.requires_grad_()

b
```

```
# Manually create gradients
x = torch.ones(2,requires_grad= True)
print(x)

y = 5 * (x + 1)**2

o = (1/2)*torch.sum(y)
o
```

```
o.backward()
x.grad
```
