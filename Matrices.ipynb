##### Seeds

Seeds are used to reproduce randomly valued results.Seeds allow to repeat exact outputs in random values

```python
np.random.seed(0)
np.random.rand(2, 4)
```

```python
torch.manual_seed(0)
torch.rand((2,4))
```
