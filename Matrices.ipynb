##### Seeds

Seeds are used to reproduce randomly valued results.Seeds allow to repeat exact outputs in random values

```python
np.random.seed(0)
np.random.rand(2, 4)
```

```python
torch.manual_seed(0)
torch.rand((2,4))
```

```python
# Fixing seed for GPU
if torch.cuda.is_available():
    torch.cuda.manual_seed_all()
```

```python
a = torch.ones(2,2)
b = torch.rand((2,2))
c = a+b

# Alternative method
c = torch.add(a,b)

# Doing it inplace to save the day
b.add_(a)
```

```python
a = torch.ones(2,2)
b = torch.rand((2,2))
c = a - b

c = torch.sub(a,b)

c.sub_(a)
```

```python
a = torch.ones(2,2)
b = torch.rand((2,1))

c =a*b

c = torch.mul(a,b)

c.mul_(a)
```

```python
a = torch.ones(2,2)
b = torch.rand((2,2))

c = b/a

c = torch.div(b,a)

c.div_(b)
```
